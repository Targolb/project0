```python
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    expression = (n**5) + (3*n) + 1
    expression = np.where(expression >= 0, expression, np.nan)  # Avoid negatives it cause a waning I want to avoid
    return np.sqrt(expression)

def g(n):
    return 5 * (n**2)

n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, f(n1), label=r'$f(n) = \sqrt{n^5 + 3n + 1}$')
plt.plot(n1, g(n1), label=r'$g(n) = 5n^2$')
plt.title('Comparison of f(n) and g(n) for n ranging from 1 to 10')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, f(n2), label=r'$f(n) = \sqrt{n^5 + 3n + 1}$')
plt.plot(n2, g(n2), label=r'$g(n) = 5n^2$')
plt.title('Comparison of f(n) and g(n) for n ranging from 1 to $10^6$')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    expression = (n**5) + (3*n) + 1
    expression = np.where(expression >= 0, expression, np.nan)  # Avoid negatives it cause a waning I want to avoid
    return np.sqrt(expression)

def g(n):
    return 5 * (n**2)

# Define the ratio f(n) / g(n)
def ratio(n):
    return f(n) / g(n)

# Define the range of n for the two plots
n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, ratio(n1), label=r'$\frac{f(n)}{g(n)}$')
plt.title(r'Ratio $\frac{f(n)}{g(n)}$ for n ranging from 1 to 10')
plt.xlabel('n')
plt.ylabel(r'$\frac{f(n)}{g(n)}$')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, ratio(n2), label=r'$\frac{f(n)}{g(n)}$')
plt.title(r'Ratio $\frac{f(n)}{g(n)}$ for n ranging from 1 to $10^6$')
plt.xlabel('n')
plt.ylabel(r'$\frac{f(n)}{g(n)}$')
plt.legend()
plt.grid(True)
plt.show()
```
