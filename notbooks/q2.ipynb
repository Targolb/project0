```python
import numpy as np
import matplotlib.pyplot as plt

def ratio(n):
    f_n = 0.5 * n * (n - 1) + 10
    g_n = n**2
    return f_n / g_n

n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, ratio(n1), label=r'$\frac{f(n)}{g(n)}$')
plt.title(r'Ratio $\frac{f(n)}{g(n)}$ for n ranging from 1 to 10')
plt.xlabel('n')
plt.ylabel(r'$\frac{f(n)}{g(n)}$')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, ratio(n2), label=r'$\frac{f(n)}{g(n)}$')
plt.title(r'Ratio $\frac{f(n)}{g(n)}$ for n ranging from 1 to $10^6$')
plt.xlabel('n')
plt.ylabel(r'$\frac{f(n)}{g(n)}$')
plt.legend()
plt.grid(True)
plt.show()
```
