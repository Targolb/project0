```python
import numpy as np
import matplotlib.pyplot as plt

# Define the functions
def f(n):
    return 0.5 * n * (n - 1) + 10

def g(n):
    return n**2

n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, f(n1), label='f(n) = (1/2)n(n-1) + 10')
plt.plot(n1, g(n1), label='g(n) = n^2')
plt.title('Comparison of f(n) and g(n)')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, f(n2), label='f(n) = (1/2)n(n-1) + 10')
plt.plot(n2, g(n2), label='g(n) = n^2')
plt.title('Comparison of f(n) and g(n) for n from 1 to 10^6')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
