```python
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    return np.log(n)

def g(n):
    return np.sqrt(n)



n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, f(n1), label=r'$f(n) = \log{n}$')
plt.plot(n1, g(n1), label=r'$g(n) = \sqrt{n}$')
plt.title('Comparison of f(n) and g(n) for n ranging from 1 to 10')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, f(n2), label=r'$f(n) = \log{n}$')
plt.plot(n2, g(n2), label=r'$g(n) = \sqrt{n}$')
plt.title('Comparison of f(n) and g(n) for n ranging from 1 to $10^6$')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

```

```python
def ratio(n):
    return f(n) / g(n)

n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, ratio(n1), label=r'$\frac{f(n)}{g(n)}$')
plt.title(r'Ratio $\frac{f(n)}{g(n)}$ for n ranging from 1 to 10')
plt.xlabel('n')
plt.ylabel(r'$\frac{f(n)}{g(n)}$')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, ratio(n2), label=r'$\frac{f(n)}{g(n)}$')
plt.title(r'Ratio $\frac{f(n)}{g(n)}$ for n ranging from 1 to $10^6$')
plt.xlabel('n')
plt.ylabel(r'$\frac{f(n)}{g(n)}$')
plt.legend()
plt.grid(True)
plt.show()
```
