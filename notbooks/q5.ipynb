Part 1:

```python
import numpy as np
import matplotlib.pyplot as plt

def f(n):
    return np.log2(n)

def g(n):
    # Prevent division by zero 
    log_vals = np.log10(n)
    log_vals[log_vals == 0] = np.finfo(float).eps  # Replace zeros with a very small number
    return log_vals

n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, f(n1), label=r'$f(n) = \log_2 n$')
plt.plot(n1, g(n1), label=r'$g(n) = \log_{10} n$')
plt.title(r'Comparison of $f(n)$ and $g(n)$ for $n$ ranging from 1 to 10')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, f(n2), label=r'$f(n) = \log_2 n$')
plt.plot(n2, g(n2), label=r'$g(n) = \log_{10} n$')
plt.title(r'Comparison of $f(n)$ and $g(n)$ for $n$ ranging from 1 to 10^6')
plt.xlabel('n')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```

Part 2:

```python

def ratio(n):
    return f(n) / g(n)

# Define the range of n for the two plots
n1 = np.arange(1, 11)  # n from 1 to 10
n2 = np.arange(1, 10**6 + 1)  # n from 1 to 10^6

# Plot for n ranging from 1 to 10
plt.figure(figsize=(10, 5))
plt.plot(n1, ratio(n1), label=r'$\frac{\log_2(n)}{\log_{10}(n)}$')
plt.title(r'Ratio $\frac{\log_2(n)}{\log_{10}(n)}$ for $n$ ranging from 1 to 10')
plt.xlabel('n')
plt.ylabel(r'$\frac{\log_2(n)}{\log_{10}(n)}$')
plt.legend()
plt.grid(True)
plt.show()

# Plot for n ranging from 1 to 10^6
plt.figure(figsize=(10, 5))
plt.plot(n2, ratio(n2), label=r'$\frac{\log_2(n)}{\log_{10}(n)}$')
plt.title(r'Ratio $\frac{\log_2(n)}{\log_{10}(n)}$ for $n$ ranging from 1 to 10^6')
plt.xlabel('n')
plt.ylabel(r'$\frac{\log_2(n)}{\log_{10}(n)}$')
plt.legend()
plt.grid(True)
plt.show()
```
